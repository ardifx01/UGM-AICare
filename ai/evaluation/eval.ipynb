# Answer Generation


```python
with open("evaluation_dataset_20250724.json", "r", encoding="utf-8") as f:
    json_data = json.load(f)


evaluation_data = json_data.get("data", [])
```

```python
from pydantic import BaseModel
from typing import Optional, List, Literal

class EvaluationDataset(BaseModel):
    id: Optional[str] = None
    query: str
    query_lable: str
    golden_nodes: List[str]
    golden_answer: str

class FetchAnswerParams(BaseModel):
    search_method: Literal['vector', 'auto']
    graph_traversal: Literal['neighbor_expansion', 'n-shortest_path', 'auto']
    top_k: int
    max_hop: int
    limit: int

evaluations = [EvaluationDataset(**item) for item in evaluation_data]
```

```python
import time
import httpx


timeout = httpx.Timeout(6000)  # 600 seconds = 10 minutes

async def fetch_answer(item, params: FetchAnswerParams, client):
    query = item['query']

    try:
        res = await client.post(
            url     = URL_GET_ANSWER, 
            json    = {
                'query': query, 
                'graph_traversal': params.graph_traversal, 
                "max_hop": params.max_hop,  
                "limit": params.limit, 
                "top_k": params.top_k, 
                "search_method": params.search_method
                }
            )

        response = res.json()
        item['graph_traversal'] = response['data']['graph_traversal']
        item['knowledge'] = response['data']['knowledge']
        item['llm_answer'] = response['data']['answer']
        item['latency'] = response['data']['latency']

        if item["graph_traversal"] == "neighbor_expansion":
            try:
                # Extract central entities and their neighbors
                graph_entities = set()
                
                for g in response["data"].get("graph", []):
                    # Add central entity
                    if "central_name" in g:
                        graph_entities.add(g["central_name"])
                    
                    # Add neighbor entities
                    for neighbor_item in g.get("neighborhood", []):
                        neighbor = neighbor_item.get("neighbor", {})
                        if "name" in neighbor:
                            graph_entities.add(neighbor["name"])
                
                # Remove duplicates while preserving order
                item["graph_entities"] = list(graph_entities)
                
            except Exception as e:
                print(f"Failed to extract neighbor_expansion result: {e}")
                item["graph_entities"] = []

        elif item["graph_traversal"] == "n-shortest_path":
            try:
                path_entities = set()
                for path in response["data"].get("graph", []):
                    path_nodes = path.get("path_nodes", [])
                    for node in path_nodes:
                        if "name" in node:
                            path_entities.add(node["name"])
                item["graph_entities"] = list(path_entities)
            except Exception as e:
                print(f"Failed to extract n-shortest_path result: {e}")
                item["graph_entities"] = []

        print(f"[✓] '{query}' latency: {item['latency']}s")

    except httpx.RequestError as e:
        print(f"[✗] '{query}' FAILED - {e}")
        item['graph_traversal'] = None
        item['knowledge'] = None
        item['llm_answer'] = None
        item['graph_entities'] = []
        item['latency'] = None

async def process_all_sequential(evaluation_data, params: FetchAnswerParams):
    async with httpx.AsyncClient(timeout=timeout) as client:
        for item in evaluation_data:
            await fetch_answer(item=item, params=params, client=client)
```

```python
from typing import List, Dict

default = FetchAnswerParams(search_method='auto', graph_traversal='auto', top_k=3, max_hop=10, limit=50)
search_vector = FetchAnswerParams(search_method='vector', graph_traversal='auto', top_k=5, max_hop=10, limit=50)
graph_traversal_neighbor_expansion = FetchAnswerParams(search_method='auto', graph_traversal='neighbor_expansion', top_k=3, max_hop=10, limit=50)
graph_traversal_n_shortest_path = FetchAnswerParams(search_method='auto', graph_traversal='n-shortest_path', top_k=3, max_hop=10, limit=50)

treatment: List[Dict[str, FetchAnswerParams]] = [
    {"name": "default", "params": default},
    {"name": "search_vector", "params": search_vector},
    {"name": "neighbor_expansion", "params": graph_traversal_neighbor_expansion},
    {"name": "n-shortest_path", "params": graph_traversal_n_shortest_path},
]
```

```python
import time
import json

for item in treatment[2:3]:
    start_all = time.perf_counter()

    await process_all_sequential(evaluation_data, params=item['params'])

    end_all = time.perf_counter()
    elapsed = end_all - start_all
    print(f"{item['name']} took {elapsed:.2f} seconds")

    # Save to separate result file per strategy for clarity
    filename = f"query_result_{item['name']}_20250724.json"
    with open(filename, "w", encoding="utf-8") as f:
        json.dump(evaluation_data, f, indent=2, ensure_ascii=False)
```
