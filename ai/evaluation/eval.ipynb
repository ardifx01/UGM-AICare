## Answer Generation For Graph Based RAG


```python
from pydantic import BaseModel
from typing import Optional, List, Literal

class EvaluationDataset(BaseModel):
    id: Optional[str] = None
    query: str
    query_lable: str
    golden_nodes: List[str]
    golden_answer: str

class FetchAnswerParams(BaseModel):
    search_method: Literal['vector', 'auto']
    graph_traversal: Literal['neighbor_expansion', 'n-shortest_path', 'auto']
    top_k: int
    max_hop: int
    limit: int

evaluations = [EvaluationDataset(**item) for item in evaluation_data]
```

```python
import time
import httpx


timeout = httpx.Timeout(6000)  # 600 seconds = 10 minutes

async def fetch_answer(item, params: FetchAnswerParams, client):
    query = item['query']

    try:
        res = await client.post(
            url     = URL_GET_ANSWER, 
            json    = {
                'query': query, 
                'graph_traversal': params.graph_traversal, 
                "max_hop": params.max_hop,  
                "limit": params.limit, 
                "top_k": params.top_k, 
                "search_method": params.search_method
                }
            )

        response = res.json()
        item['graph_traversal'] = response['data']['graph_traversal']
        item['knowledge'] = response['data']['knowledge']
        item['llm_answer'] = response['data']['answer']
        item['latency'] = response['data']['latency']

        if item["graph_traversal"] == "neighbor_expansion":
            try:
                # Extract central entities and their neighbors
                graph_entities = set()
                
                for g in response["data"].get("graph", []):
                    # Add central entity
                    if "central_name" in g:
                        graph_entities.add(g["central_name"])
                    
                    # Add neighbor entities
                    for neighbor_item in g.get("neighborhood", []):
                        neighbor = neighbor_item.get("neighbor", {})
                        if "name" in neighbor:
                            graph_entities.add(neighbor["name"])
                
                # Remove duplicates while preserving order
                item["graph_entities"] = list(graph_entities)
                
            except Exception as e:
                print(f"Failed to extract neighbor_expansion result: {e}")
                item["graph_entities"] = []

        elif item["graph_traversal"] == "n-shortest_path":
            try:
                path_entities = set()
                for path in response["data"].get("graph", []):
                    path_nodes = path.get("path_nodes", [])
                    for node in path_nodes:
                        if "name" in node:
                            path_entities.add(node["name"])
                item["graph_entities"] = list(path_entities)
            except Exception as e:
                print(f"Failed to extract n-shortest_path result: {e}")
                item["graph_entities"] = []

        print(f"[✓] '{query}' latency: {item['latency']}s")

    except httpx.RequestError as e:
        print(f"[✗] '{query}' FAILED - {e}")
        item['graph_traversal'] = None
        item['knowledge'] = None
        item['llm_answer'] = None
        item['graph_entities'] = []
        item['latency'] = None

async def process_all_sequential(evaluation_data, params: FetchAnswerParams):
    async with httpx.AsyncClient(timeout=timeout) as client:
        for item in evaluation_data:
            await fetch_answer(item=item, params=params, client=client)
```

```python
import numpy as np

# --- Metric Functions (copy from above) ---
def calculate_precision(retrieved_items: list, golden_items: set) -> float:
    retrieved_set = set(retrieved_items)
    relevant_retrieved = retrieved_set.intersection(golden_items)
    if not retrieved_set: return 0.0
    return len(relevant_retrieved) / len(retrieved_set)

def calculate_recall(retrieved_items: list, golden_items: set) -> float:
    retrieved_set = set(retrieved_items)
    relevant_retrieved = retrieved_set.intersection(golden_items)
    if not golden_items: return 1.0
    return len(relevant_retrieved) / len(golden_items)

def calculate_reciprocal_rank(retrieved_items: list, golden_items: set) -> float:
    for i, item in enumerate(retrieved_items):
        if item in golden_items:
            return 1.0 / (i + 1)
    return 0.0

def evaluate_retrieval_system(results, golden):
    precisions = []
    recalls = []
    reciprocal_ranks = []
    failed = []

    for query_id, retrieved in results.items():
        if query_id in golden:
            golden_set = golden[query_id]

            precisions.append(calculate_precision(retrieved, golden_set))
            recalls.append(calculate_recall(retrieved, golden_set))
            reciprocal_rank = calculate_reciprocal_rank(retrieved, golden_set)
            reciprocal_ranks.append(reciprocal_rank)

            if(reciprocal_rank <= 0):
                failed.append(query_id)

    # Calculate average metrics
    avg_precision = np.mean(precisions) if precisions else 0.0
    avg_recall = np.mean(recalls) if recalls else 0.0
    mrr = np.mean(reciprocal_ranks) if reciprocal_ranks else 0.0

    hit_ratio = 1- (len(failed)/len(reciprocal_ranks))

    return {
        "average_precision": avg_precision,
        "average_recall": avg_recall,
        "mrr": mrr,
        "hit_ratio": hit_ratio
    }
```

```python
# RAGAS Rate Limit Handler
# Specialized script to handle OpenAI rate limits gracefully

import time
import json
import pandas as pd
import math
import re
from datasets import Dataset
from ragas import evaluate
from ragas.metrics import answer_similarity, answer_correctness, faithfulness, answer_relevancy
from ragas.llms.base import ChatOpenAI, OpenAI
from ragas.embeddings.base import OpenAIEmbeddings

# from ragas.llms.openai import OpenAILLM
# from ragas.embeddings.openai import OpenAIEmbeddings
import os

class RateLimitedEvaluator:
    """
    A wrapper class for RAGAS evaluation that handles rate limits intelligently
    """
    
    def __init__(self, api_key, model="gpt-4.1-nano", max_tokens=5000):
        self.api_key = api_key
        self.model = model
        self.max_tokens = max_tokens
        self.setup_llm()
        
    def setup_llm(self):
        """Initialize LLM with rate-limit friendly settings"""
        try:
            self.llm = ChatOpenAI(
                model=self.model,
                api_key=self.api_key,
                temperature=0.1,
                max_tokens=self.max_tokens
            )
            
            self.embeddings = OpenAIEmbeddings(
                model="text-embedding-ada-002",
                api_key=self.api_key
            )
            
            print(f"✅ Initialized {self.model} with rate limit handling")
            
        except Exception as e:
            print(f"❌ Failed to initialize LLM: {str(e)}")
            raise
    
    def extract_wait_time(self, error_message):
        """Extract wait time from OpenAI rate limit error"""
        # Look for "try again in X.Xs" pattern
        wait_match = re.search(r'try again in (\d+\.?\d*)s', error_message)
        if wait_match:
            return float(wait_match.group(1)) + 2  # Add 2 second buffer
        
        # Default fallback wait times based on error type
        if "rate_limit" in error_message.lower():
            return 30  # 30 seconds for rate limits
        elif "429" in error_message:
            return 60  # 1 minute for 429 errors
        else:
            return 10  # 10 seconds for other errors
    
    def evaluate_batch(self, batch_dataset, metrics, max_retries=5):
        """Evaluate a single batch with retry logic"""
        
        for attempt in range(max_retries):
            try:
                print(f"      Attempt {attempt + 1}/{max_retries}...")
                
                result = evaluate(
                    dataset=batch_dataset,
                    metrics=metrics,
                    llm=self.llm,
                    embeddings=self.embeddings,
                    raise_exceptions=True
                )
                
                return result.to_pandas(), True  # Success
                
            except Exception as e:
                error_str = str(e)
                print(f"      ❌ Attempt {attempt + 1} failed: {error_str[:80]}...")
                
                # Check if it's a rate limit error
                if any(keyword in error_str.lower() for keyword in ['rate_limit', '429', 'quota']):
                    wait_time = self.extract_wait_time(error_str)
                    print(f"      ⏳ Rate limited. Waiting {wait_time:.1f} seconds...")
                    time.sleep(wait_time)
                    
                elif "timeout" in error_str.lower():
                    wait_time = 15 * (attempt + 1)  # Increasing timeout waits
                    print(f"      ⏳ Timeout error. Waiting {wait_time} seconds...")
                    time.sleep(wait_time)
                    
                else:
                    # Other errors - shorter wait
                    wait_time = 5 * (attempt + 1)
                    print(f"      ⏳ Unknown error. Waiting {wait_time} seconds...")
                    time.sleep(wait_time)
                
                if attempt == max_retries - 1:
                    print(f"      ❌ Batch failed after {max_retries} attempts")
                    return None, False
        
        return None, False
    
    def evaluate_dataset(self, dataset, metrics, batch_size=3, base_delay=10, 
                        progress_file="evaluation_progress.json"):
        """
        Evaluate entire dataset with intelligent rate limiting
        
        Args:
            dataset: RAGAS dataset to evaluate
            metrics: List of RAGAS metrics to use
            batch_size: Number of records per batch (keep small)
            base_delay: Base delay between batches in seconds
            progress_file: File to save progress (for resuming)
        """
        
        print(f"🚀 Starting intelligent rate-limited evaluation")
        print(f"📊 Dataset size: {len(dataset)} records")
        print(f"📦 Batch size: {batch_size}")
        print(f"⏱️ Base delay: {base_delay} seconds")
        print(f"📋 Metrics: {[m.name for m in metrics]}")
        
        # Load previous progress if exists
        completed_batches = 0
        all_results = []
        
        try:
            with open(progress_file, 'r') as f:
                progress = json.load(f)
                completed_batches = progress.get('completed_batches', 0)
                if progress.get('results'):
                    all_results = [pd.DataFrame(result) for result in progress['results']]
                    print(f"📂 Resuming from batch {completed_batches + 1}")
        except FileNotFoundError:
            print("📝 Starting fresh evaluation")
        
        total_batches = math.ceil(len(dataset) / batch_size)
        failed_batches = []
        
        # Calculate estimated time
        estimated_time = (total_batches - completed_batches) * (base_delay + 30)  # 30s avg per batch
        print(f"⏰ Estimated time: {estimated_time / 60:.1f} minutes")
        
        for batch_idx in range(completed_batches, total_batches):
            start_idx = batch_idx * batch_size
            end_idx = min((batch_idx + 1) * batch_size, len(dataset))
            
            print(f"\n🔄 Batch {batch_idx + 1}/{total_batches} (records {start_idx+1}-{end_idx})")
            print(f"   Progress: {(batch_idx/total_batches)*100:.1f}%")
            
            # Create batch
            batch_indices = list(range(start_idx, end_idx))
            batch_dataset = dataset.select(batch_indices)
            
            # Evaluate batch
            batch_result, success = self.evaluate_batch(batch_dataset, metrics)
            
            if success:
                all_results.append(batch_result)
                print(f"   ✅ Batch {batch_idx + 1} completed successfully!")
                
                # Save progress
                progress_data = {
                    'completed_batches': batch_idx + 1,
                    'total_batches': total_batches,
                    'failed_batches': failed_batches,
                    'results': [df.to_dict('records') for df in all_results]
                }
                
                with open(progress_file, 'w') as f:
                    json.dump(progress_data, f)
                
            else:
                failed_batches.append(batch_idx)
                print(f"   ❌ Batch {batch_idx + 1} failed permanently")
                
                # Create dummy results for failed batch
                dummy_data = {
                    'question': [item['question'] for item in batch_dataset],
                    'answer': [item['answer'] for item in batch_dataset],
                    'contexts': [item['contexts'] for item in batch_dataset],
                    'ground_truth': [item['ground_truth'] for item in batch_dataset]
                }
                
                # Add dummy scores
                for metric in metrics:
                    dummy_data[metric.name] = [0.0] * len(batch_dataset)
                
                all_results.append(pd.DataFrame(dummy_data))
            
            # Adaptive delay between batches
            if batch_idx < total_batches - 1:
                # Increase delay if we've had recent failures
                recent_failures = len([b for b in failed_batches if b >= batch_idx - 5])
                adaptive_delay = base_delay + (recent_failures * 5)
                
                print(f"   ⏳ Waiting {adaptive_delay} seconds before next batch...")
                time.sleep(adaptive_delay)
        
        # Combine results
        if all_results:
            final_results = pd.concat(all_results, ignore_index=True)
            print(f"\n✅ Evaluation completed!")
            print(f"📊 Total records processed: {len(final_results)}")
            print(f"❌ Failed batches: {len(failed_batches)}")
            
            # Clean up progress file
            try:
                os.remove(progress_file)
            except:
                pass
            
            return final_results
        else:
            print("\n❌ No results obtained")
            return None

# Usage example
def run_evaluation(json_file_path, api_key, output_file="rate_limited_results.csv"):
    """
    Complete evaluation pipeline with rate limiting
    """
    
    # Load data
    print("📂 Loading data...")
    with open(json_file_path, 'r', encoding='utf-8') as f:
        query_result = json.load(f)
    
    # Convert to DataFrame
    df = pd.DataFrame([{
        'query': item.get('query', ''),
        'golden_answer': item.get('golden_answer', ''),
        'context': item.get('context', ''),
        'llm_answer': item.get('llm_answer', '')
    } for item in query_result])
    
    print(f"📊 Loaded {len(df)} records")
    
    # Convert to RAGAS format
    ragas_data = {
        'question': df['query'].tolist(),
        'answer': df['llm_answer'].tolist(),
        'contexts': [[str(context)] if context else [""] for context in df['context']],
        'ground_truth': df['golden_answer'].tolist()
    }

    ragas_dataset = Dataset.from_dict(ragas_data)
    
    # Initialize evaluator
    evaluator = RateLimitedEvaluator(api_key=api_key, model="gpt-4.1-nano")
    
    # Select metrics (keep minimal to reduce API calls)
    metrics = [answer_similarity, answer_correctness, answer_relevancy, faithfulness]
    
    # Run evaluation
    results = evaluator.evaluate_dataset(
        dataset=ragas_dataset,
        metrics=metrics,
        batch_size=5,      
        base_delay=2
    )
    
    if results is not None:
        # Save results
        results.to_csv(output_file, index=False)
        print(f"💾 Results saved to {output_file}")
        
        # Show summary
        print(f"\n📈 Summary:")
        for metric in metrics:
            if metric.name in results.columns:
                avg_score = results[metric.name].mean()
                print(f"   {metric.name}: {avg_score:.4f}")
        
        return results
    else:
        print("❌ Evaluation failed completely")
        return None
```
